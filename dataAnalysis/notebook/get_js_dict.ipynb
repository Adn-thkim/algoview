```python
# 문제유형별

def get_js_p_type_p_name(df):
    '''
    JavaScript 문제의 문제유형별 문제 수를 카운트하는 함수입니다.
    '''

    # 파일명 중복 제거
    df = df.drop_duplicates(['file_name'], keep='first')

    # 문제유형별 count
    df = df['problem_type'].value_counts().to_frame()

    return df


def get_js_p_type_method(df):
    '''
    JavaScript 문제의 문제유형마다 사용된 메소드별 횟수를 카운트하는 함수입니다.
    '''

    # 예외 처리 : 메소스 사용횟수 0인 행 삭제
    df = df[df["function_method"] != 0]

    # 문제유형별 메소드 사용횟수 추출
    df = df.groupby(['problem_type', 'function_method'], as_index=False)['countmethod'].sum()

    # 행렬 모양으로 변환 (pivot)
    df = df.pivot(index='problem_type', columns='function_method', values='countmethod')

    return df
```

```python
def get_js_dict():
    '''
    JavaScript 문제의 전체 데이터를 dictionary 형태로 변환하여 반환하는 함수입니다.
    '''

    js_df = json_to_dataframe("js")
    
    # 전체 데이터 저장할 딕셔너리
    js_dict = dict()


    # 딕셔너리에 파트별 데이터 삽입

    
    js_dict["js_p_type_p_name"] = get_js_p_type_p_name(js_df).to_dict()["problem_type"]
    js_dict["js_p_type_method"] = get_js_p_type_method(js_df).to_dict()


    return js_dict
```
